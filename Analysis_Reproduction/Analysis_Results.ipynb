# Person-independent F1-score results for state-of-the-art self-supervised methods on DISFA. Higher is better for F1-score.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import pickle

with open('../Results/Bootstraps_results_Figure2.pickle', 'rb') as handle:
    F1_scores_dict = pickle.load(handle)
```

## Results are averaged across AUs

```python

sns.set(style="whitegrid")

palette ={"FaceCycle": "olivedrab",
          "TCAE": "lightsalmon",
          "temporal-consistency": "thistle",
          "FAb-Net": "lightsteelblue"}

dataframe = pd.DataFrame({"Model":np.concatenate([([i]*300) for i in 
                         ["FaceCycle","TCAE","temporal-consistency","FAb-Net"]], axis=0),
                         "avg_f1_score":0})

dataframe.loc[dataframe.Model == "FaceCycle","avg_f1_score"] = F1_scores_dict["avg_F1_Bootstrap_FaceCycle"]
dataframe.loc[dataframe.Model == "TCAE","avg_f1_score"] = F1_scores_dict["avg_F1_Bootstrap_TCAE"]
dataframe.loc[dataframe.Model == "temporal-consistency","avg_f1_score"] = F1_scores_dict["avg_F1_Bootstrap_temporal-consistency"]
dataframe.loc[dataframe.Model == "FAb-Net","avg_f1_score"] = F1_scores_dict["avg_F1_Bootstrap_FAb-Net"]




x = "Model"
y = "avg_f1_score"

plt.figure(figsize=(12,8))
ax = sns.barplot(data=dataframe, x=x, y=y,palette=palette)
plt.xlabel("Models")
plt.ylabel("3-cross validation Person-dependent F1 score")
plt.ylim(0.25,0.38)

plt.show()

print("Mean:","\n", 
      "FaceCycle", dataframe.loc[dataframe.Model == "FaceCycle","avg_f1_score"].mean(),"\n",
      "TCAE", dataframe.loc[dataframe.Model == "TCAE","avg_f1_score"].mean(),"\n",
      "temporal-consistency", dataframe.loc[dataframe.Model == "temporal-consistency","avg_f1_score"].mean(),"\n",
      "FAb-Net", dataframe.loc[dataframe.Model == "FAb-Net","avg_f1_score"].mean())
```

## Results are displayed per AU

```python
AUs_to_use = ['au'+str(i) for i in [1,2,4,6,9,12,25,26]]
sns.set(style="white")
palette ={"FaceCycle": "olivedrab",
          "TCAE": "lightsalmon",
          "temporal-consistency": "thistle",
          "FAb-Net": "lightsteelblue"}

df = pd.DataFrame({
    "Model":np.concatenate([([i]*len(AUs_to_use)*300) for i in 
                           ["FaceCycle","TCAE","temporal-consistency","FAb-Net"]], axis=0),
                           "avg_f1_score":0,
                           "AUs":AUs_to_use*300*4})

df.loc[df.Model=="FaceCycle","avg_f1_score"] = [item for sublist in F1_scores_dict["F1_Bootstrap_FaceCycle"] 
                                                for item in sublist]
df.loc[df.Model=="TCAE","avg_f1_score"] = [item for sublist in F1_scores_dict["F1_Bootstrap_TCAE"] 
                                           for item in sublist]
df.loc[df.Model=="temporal-consistency","avg_f1_score"] = [item for sublist in 
                                                           F1_scores_dict["F1_Bootstrap_temporal-consistency"] 
                                                           for item in sublist]
df.loc[df.Model=="FAb-Net","avg_f1_score"] = [item for sublist in F1_scores_dict["F1_Bootstrap_FAb-Net"] 
                                              for item in sublist]


plt.figure(figsize=(20,10))
graph=sns.barplot(x="AUs", hue="Model", y="avg_f1_score", data=df,palette=palette)
for i in range(len(AUs_to_use)):
    graph.axvline(0.5+i,color="black",alpha=0.3)
graph.axhline(1,color="black")    
plt.legend('')
plt.ylabel("3-cross-validation Person independent F1 score per action units")
plt.ylim(0,0.8)
plt.show()


```

## Clustering analysis in AU space from the learned embedding by GM and PSM and discovery of novel meaningful clusters

```python
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
from collections import Counter
import pandas as pd
import numpy as np
import pickle

AUs= ["au"+str(i) for i in [1,2,4,5,6,9,12,15,17,20,25,26]]

colors_aus = dict()
colors_aus['au1'] = "blue"
colors_aus['au2'] = "green"
colors_aus['au4'] = "red"
colors_aus['au5'] = "grey"
colors_aus['au6'] = "black"
colors_aus['au9'] = "yellow"
colors_aus['au12'] = "cyan"
colors_aus['au15'] = "magenta"
colors_aus['au17'] = "brown"
colors_aus['au20'] = "purple"
colors_aus['au25'] = "lavender"
colors_aus['au26'] = "rosybrown"
colors_aus['None'] = "mistyrose"

with open('../Results/clusters_Figure4.pickle', 'rb') as handle:
    dict_clusters = pickle.load(handle)
```

```python

sns.set_style("white")

different_PSM_clusters=[]
different_GM_clusters=[]

for SN in ["SN004","SN018"]:
    print(SN)
    for model_type in ["GM","PSM"]:        
        df = dict_clusters[SN][model_type]
        ncols= len(df.label.unique()) # one column per cluster
        
        count=0  
        fig,axs = plt.subplots(nrows=1,ncols=ncols,figsize=(ncols*2,2))
        sum_AUs_per_label = [[df[df.label==i][au].sum()/len(df.loc[df.label==i])
                              for i in sorted(df["label"].unique())] for au in [l+"_bis" for l in AUs]]
        sum_AUs_per_label_dict = { AUs[i] : sum_AUs_per_label[i] for i in range(len(AUs)) }
        if model_type=="GM":
            df_AU_GM = pd.DataFrame(sum_AUs_per_label_dict, index=sorted(df["label"].unique()))
            tmp = df_AU_GM
        else:
            df_AU_PSM = pd.DataFrame(sum_AUs_per_label_dict, index=sorted(df["label"].unique()))
            tmp = df_AU_PSM
        for cluster in tmp.index:
            tmp.loc[cluster].plot.bar(stacked=False,color=list(colors_aus.values()),
                                           title="Cluster"+str(cluster),legend=False,ax=axs[count])
            if count>0:
                axs[count].yaxis.set_ticklabels([])
            axs[count].set_ylim(0,1)
            count+=1
        plt.suptitle(model_type+" "+str(Counter(df.label)),y=1.15)
        plt.show()
    

    
    # Prepare PSM and GM data for correlation matrices
    
    df_PSM_to_use = df_AU_PSM.copy().T
    df_PSM_to_use = df_PSM_to_use.loc[:, df_PSM_to_use.any()]
    df_GM_to_use = df_AU_GM.copy().T
    df_GM_to_use = df_GM_to_use.loc[:, df_GM_to_use.any()]
    df_PSM_to_use.columns = ["PSM_"+str(col)  for col in df_PSM_to_use.columns]
    df_GM_to_use.columns = ["GM_"+str(col)  for col in df_GM_to_use.columns]

    # Prepare correlation matrix    
    df_concat = pd.concat([df_PSM_to_use,df_GM_to_use],axis=1)
    df_concat_corr = df_concat.corr().loc[
            [l for l in df_concat.columns if "PSM" in l],[l for l in df_concat.columns if "GM" in l]]

    # Prepare euclidean distance matrix
    df_dist = pd.DataFrame(columns= df_GM_to_use.columns, index=df_PSM_to_use.columns)
    for ind in df_dist.index:
        df_dist.loc[ind] = [np.abs(df_concat[ind]-df_concat[gm_col]).sum() for gm_col in df_dist.columns]
    df_dist = df_dist.astype("float")

    
    # Metric = correlation - euclidean distance

    new_metric_df = df_concat_corr-df_dist
    max_val = new_metric_df.max().max()
    min_val = new_metric_df.min().min()
    new_metric_df = (new_metric_df-min_val)/(max_val-min_val)
    new_metric_df = new_metric_df.astype("float")
    
    # Plot metric heatmap
    fig, ax = plt.subplots(ncols=1,figsize=(5,3))
    sns.heatmap(new_metric_df,ax=ax)
    ax.set_title("Metric = Correlation - Euclidean Distance")
    plt.show()
    
    # Plot matrix with metric values
    print(tabulate(new_metric_df, headers='keys', tablefmt='psql'))
    
    # Detect novel clusters with threshold 0.8:
    new_PSM = (new_metric_df > 0.8).any(axis=1)==False
    new_GM = (new_metric_df > 0.8).any(axis=0)==False   
    print("Novel Clusters from PSM :",new_PSM[new_PSM].index.values.tolist())
    print("Novel Clusters from GM :",new_GM[new_GM].index.values.tolist())
    different_PSM_clusters.append(new_PSM[new_PSM].index.values.tolist())
    different_GM_clusters.append(new_GM[new_GM].index.values.tolist())
    
    



    
```

# Quantitative and qualitative effects of transfer Learning 

## Person-dependent F1-score results for comparing the transfer learning approaches ((2) and (3)) to the upper and lower performance bounds ((1) and (4)). 

### Results are averaged across AUs and individuals. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import pickle
# Store data (serialize)
with open('../Results/Bootstrap_F1_scores_Training_Approaches_Figure5.pickle', 'rb') as handle:
    All_data_results_f1_score = pickle.load(handle)
```

```python
# We need to prepare a dictionary for the different training approaches:

# (1) Approach 1: PSM Trained 500 epochs on S1: We trained one PSM on test person for 500 epochs,
#                 same as in Figure 3.

# (2) Approach 2: Pretrained VoxCeleb + PSM Trained 10 epochs on S1: We used transfer learning. 
#     We used the pretrained facial-motion cycle consistency on VoxCeleb across persons and resumed 
#     the training for 10 epochs on the test person. 
#     We computed the F1 score results on all test persons and averaged the results.

# (3) Approach 3: Pretrained PSM S2 + PSM Trained 10 epochs on S1: We used transfer learning. 
#     We pretrained PSM on person 2 (any other person different than test person)
#     for 500 epochs and used the PSM pretrained model to train on the test person for 10 epochs. 
#    We computed the F1 score results on all possible train/test persons and averaged the results.

# (4) Approach 4: PSM Trained 10 epochs S1: We trained one PSM on test person for 10 epochs. 


SNs = ['SN001','SN002','SN003','SN004','SN005','SN006','SN007','SN008','SN009',
       'SN010','SN011','SN012','SN013','SN016','SN017','SN018','SN021','SN023',
       'SN024','SN025','SN026','SN027','SN028','SN029','SN030','SN031','SN032']

# Create dictionary summarizing the F1 scores of the different approaches
dict_Bootstrap_F1_scores_Training_Approaches = dict()


# Aggregate all F1 scores from approach 1.

dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"] = []
for dataset in [l for l in All_data_results_f1_score.keys() if "NO_pretraining_trained_ALL_FRAMES_on_that_subject_for_500_epochs" in l]:
    dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"].append(All_data_results_f1_score[dataset])
    
dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"]  = [x for xs in 
                                                              dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"]
                                                              for x in xs]
dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"]  = [x for xs in 
                                                              dict_Bootstrap_F1_scores_Training_Approaches["Approach 1"]
                                                              for x in xs]

# Aggregate all F1 scores from approach 2.

dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"] = []
for dataset in [l for l in All_data_results_f1_score.keys() if "VoxCeleb" in l]:
#     print(dataset)
    dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"].append(All_data_results_f1_score[dataset])
    
dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"] = [x for xs in 
                                                             dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"]
                                                             for x in xs]
dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"] = [x for xs in 
                                                             dict_Bootstrap_F1_scores_Training_Approaches["Approach 2"]
                                                             for x in xs]

# Aggregate all F1 scores from approach 3 by averaging for each Person.

dict_Bootstrap_F1_scores_Training_Approaches["Approach 3"] = []
for SN in SNs:
    list_pretrained = [l for l in SNs if l!=SN]
    tmp_list = [] # To average for each au value
    for SN_pretrained in list_pretrained:
        tmp_list.append([x for xs in All_data_results_f1_score[SN+"_pretrained_on_"+SN_pretrained] for x in xs])
        
    dict_Bootstrap_F1_scores_Training_Approaches["Approach 3"].append(np.average(np.array(tmp_list),axis=0).tolist())

dict_Bootstrap_F1_scores_Training_Approaches["Approach 3"] = [x for xs in 
                                                             dict_Bootstrap_F1_scores_Training_Approaches["Approach 3"]
                                                             for x in xs]

# Aggregate all F1 scores from approach 4.

dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"] = []
for dataset in [l for l in All_data_results_f1_score.keys() if 
                "NO_pretraining_trained_486_FRAMES_on_that_subject_for_10_epochs" in l]:
    dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"].append(All_data_results_f1_score[dataset])

dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"] = [x for xs in 
                                                             dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"]
                                                             for x in xs]
dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"] = [x for xs in 
                                                             dict_Bootstrap_F1_scores_Training_Approaches["Approach 4"]
                                                             for x in xs]
```
